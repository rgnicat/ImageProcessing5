```python
height , width = 150,210
flag_russia = np.zeros((height, width , channel),dtype=np.uint8)
flag_russia[:50,:,:] = [255,255,255]
flag_russia[50:100,:,:] = [0,0,255]
flag_russia[100:,:,:] = [255,0,0]
```

```python

plt.imshow(flag_russia)
plt.axis("off")
plt.title("RUSSIA image")
```

```python

plt.imshow(flag_russia)
plt.axis("off")
plt.title("RUSSIA image")
# --- RV LAV ET ---
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_edgecolor("black")   # riv rngi
    spine.set_linewidth(8)         # qalnlq

plt.show()
```

```python
import cv2

bordered = cv2.copyMakeBorder(flag_russia, 2, 2, 2, 2, cv2.BORDER_CONSTANT, value=[0, 0, 0])
plt.imshow(bordered)
plt.axis("off")
plt.show()
```
